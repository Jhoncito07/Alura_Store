### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
#Imprimimos una lista de las columnas de las tiendas
```

```python
print(tienda.columns)
print(tienda2.columns)
print(tienda3.columns)
print(tienda4.columns)
```

#1. Análisis de facturación



```python
#Definir el ingreso total por cada Tienda
```

```python
ingreso_total1 = (tienda['Precio']/100).sum()
ingreso_total2 = (tienda2['Precio']/100).sum()
ingreso_total3 = (tienda3['Precio']/100).sum()
ingreso_total4 = (tienda4['Precio']/100).sum()
```

```python
#Imprimir Ingreso total por cada tienda
```

```python
print(f"Facturación total de Tienda 1: ${ingreso_total1:,.2f}")
print(f"Facturación total de Tienda 2: ${ingreso_total2:,.2f}")
print(f"Facturación total de Tienda 3: ${ingreso_total3:,.2f}")
print(f"Facturación total de Tienda 4: ${ingreso_total4:,.2f}")
```

# 2. Ventas por categoría

```python
#Definimos ventas x categoria para cada tienda
```

```python
ventasXcategoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventasXcategoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventasXcategoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventasXcategoria_tienda4 = tienda4['Categoría del Producto'].value_counts()
```

```python
#Imprimimos las ventas x categoria en orden
```

```python
print("Ventas por categoría de Tienda 1:")
print(ventasXcategoria_tienda1)
print("\nVentas por categoría de Tienda 2:")
print(ventasXcategoria_tienda2)
print("\nVentas por categoría de Tienda 3:")
print(ventasXcategoria_tienda3)
print("\nVentas por categoría de Tienda 4:")
print(ventasXcategoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
#Definimos el promedio de las calificaciones para cada tienda
```

```python
promedio_calificacion_tienda1 = tienda['Calificación'].mean()
promedio_calificacion_tienda2 = tienda2['Calificación'].mean()
promedio_calificacion_tienda3 = tienda3['Calificación'].mean()
promedio_calificacion_tienda4 = tienda4['Calificación'].mean()
```

```python
#Imprimimos el promedio de las calificaciones
```

```python
print('La calificación Promedio de la tienda1 es de: {:.2f}'.format(promedio_calificacion_tienda1))
print('La calificación Promedio de la tienda2 es de: {:.2f}'.format(promedio_calificacion_tienda2))
print('La calificación Promedio de la tienda3 es de: {:.2f}'.format(promedio_calificacion_tienda3))
print('La calificación Promedio de la tienda4 es de: {:.2f}'.format(promedio_calificacion_tienda4))
```

# 4. Productos más y menos vendidos

```python
#definimos la función de producto mas vendidos
```

```python
def producto_mas_vendido(list):
    producto_mas_vendido = list['Producto'].value_counts().head()
    return producto_mas_vendido
```

```python
#definimos la función de producto menos vendidos
```

```python
def producto_menos_vendido(list):
    producto_menos_vendido = list['Producto'].value_counts().tail()
    return producto_menos_vendido
```

```python
#Imprimimos los Productos mas y menos vendidos en cada tienda
```

```python
print("Tienda 1")
print("Producto más vendido:", producto_mas_vendido(tienda))
print("\nProducto menos vendido:", producto_menos_vendido(tienda))
print('----------------------------------------------------')
print("\nTienda 2")
print("Producto más vendido:", producto_mas_vendido(tienda2))
print("\nProducto menos vendido:", producto_menos_vendido(tienda2))
print('----------------------------------------------------')
print("\nTienda 3")
print("Producto más vendido:", producto_mas_vendido(tienda3))
print("\nProducto menos vendido:", producto_menos_vendido(tienda3))
print('----------------------------------------------------')
print("\nTienda 4")
print("Producto más vendido:", producto_mas_vendido(tienda4))
print("\nProducto menos vendido:", producto_menos_vendido(tienda4))
```

# 5. Envío promedio por tienda

```python
#definimos la función del Promedio de costo del envio
```

```python
def envio_promedio(list):
    envio_promedio = (list['Costo de envío']/100).mean()
    return envio_promedio
```

```python
#Imprimimos el Promedio de costo del envio para cada tienda
```

```python
print("Tienda 1")
print(f"Envío promedio: ${envio_promedio(tienda):.2f}")
print("\nTienda 2")
print(f"Envío promedio: ${envio_promedio(tienda2):.2f}")
print(f"\nTienda 3")
print(f"Envío promedio: ${envio_promedio(tienda3):.2f}")
print(f"\nTienda 4")
print(f"Envío promedio: ${envio_promedio(tienda4):.2f}")
```

# **6. Visualización de Datos**

```python
import matplotlib.pyplot as plt
```

```python
#FACTURACIÓN POR TIENDA
```

```python
#Lista con nombres de las 4 tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
```

```python
#Grafica para Facturaciones x Tienda
facturaciones = [ingreso_total1, ingreso_total2, ingreso_total3, ingreso_total4]

#Crear grafica de barras y mostrarla
plt.bar(tiendas, facturaciones, color="skyblue")
plt.xlabel('Tiendas')
plt.ylabel('Facturación')
plt.title('Facturación por Tienda')
plt.show()
```

Este gráfico permitió identificar que la **Tienda 1** tiene la mayor facturación, lo que indica un volumen de ventas alto. En contraste, la **Tienda 4** obtuvo la facturación más baja, siendo la menos rentable en términos de ingresos totales.




---



```python
#Ventas x Categoría en cada Tienda
```

```python
df_ventas_categoria = pd.concat(
    [ventasXcategoria_tienda1, ventasXcategoria_tienda2, ventasXcategoria_tienda3, ventasXcategoria_tienda4],
    axis=1
)

df_ventas_categoria.columns = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Crear gráfico de barras agrupadas y mostrar grafica
df_ventas_categoria.plot(kind="bar", figsize=(10,6))
plt.title("Ventas por Categoría en las 4 Tiendas")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)
plt.legend(title="Tienda")
plt.tight_layout()
plt.show()
```

A través de este gráfico se observa que **Muebles** es la categoría más vendida en las cuatro tiendas.

La **Tienda 3** lidera en ventas de esta categoría con 499 unidades. Las categorías menos vendidas fueron distintas por tienda, destacando que los **Artículos para el hogar** fueron los de menor salida en las tiendas 1, 2 y 3, mientras que en la tienda 4 lo fue **Instrumentos musicales**.



---



```python
#Calificación Promedio x Tienda
```

```python
Calificacion_promedio = [promedio_calificacion_tienda1, promedio_calificacion_tienda2, promedio_calificacion_tienda3, promedio_calificacion_tienda4]

#Crear grafico circular y mostrar la grafica

plt.pie(Calificacion_promedio, labels=tiendas, autopct='%1.1f%%')
plt.title('Calificación Promedio por Tienda')
plt.show()
```

El gráfico de calificaciones muestra que la **Tienda 3** recibió la mejor valoración promedio, lo que sugiere mayor satisfacción por parte de los clientes.

La **Tienda 1** fue la peor calificada, lo que puede estar vinculado a problemas de atención, calidad o experiencia de compra.




---



```python
#Productos mas y menos Vendidos x Tienda
```

```python
Mas_y_menos_vendido = [
    producto_mas_vendido(tienda).values[0], producto_menos_vendido(tienda).values[0],
    producto_mas_vendido(tienda2).values[0], producto_menos_vendido(tienda2).values[0],
    producto_mas_vendido(tienda3).values[0], producto_menos_vendido(tienda3).values[0],
    producto_mas_vendido(tienda4).values[0], producto_menos_vendido(tienda4).values[0]
]

#Crear grafico circular y mostrar la grafica
plt.pie(Mas_y_menos_vendido, labels=tiendas*2, autopct="%1.1f%%", colors=["blue", "green", "red", "purple"]*2, startangle=90)
plt.title("Distribución de los Productos Más y Menos Vendidos por Tienda")
plt.show()
```

En esta gráfica observamos que:

- **La Tienda 3** destaca con el mayor porcentaje combinado (**16.4%** y **15.7%**), lo que indica que tiene tanto un producto muy popular como uno poco demandado con alto volumen relativo. Esto puede reflejar un mayor rango de ventas y variedad en el comportamiento de sus productos.
- **La Tienda 1** también muestra una presencia significativa con **15.2%** y **14.4%**, lo que indica un desempeño relativamente estable en términos de volumen de sus productos más y menos vendidos.
- **La Tienda 2** se mantiene en un rango medio con **9.8%** y **9.6%**, lo que podría reflejar un equilibrio pero también menor volumen comparativo.
- **La Tienda 4**, con **9.6%** y **9.3%**, muestra la menor proporción en ambos extremos, lo que podría interpretarse como un desempeño más plano, sin productos muy destacados ni muy rezagados, pero también como una señal de menor participación general en ventas destacadas.




---



```python
#Costo de envio Promedio x Tienda
```

```python
Costo_envio_promedio = [envio_promedio(tienda), envio_promedio(tienda2), envio_promedio(tienda3), envio_promedio(tienda4)]

# Crear gráfico de barras horizontales
plt.figure(figsize=(8, 5))
plt.barh(tiendas, Costo_envio_promedio, color="green", alpha=0.7)
plt.xlabel("Costo de Envío")
plt.ylabel("Tiendas")
plt.title("Costo de Envío Promedio por Tienda")
plt.grid(axis="x", linestyle="--", alpha=0.5)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

El análisis de este gráfico revela que la Tienda 1 tiene el costo de envío más alto lo que podría reducir sus márgenes de ganancia.

En cambio, la Tienda 4 tiene el costo de envío más bajo, lo cual es positivo desde una perspectiva logística, aunque no compensa su bajo desempeño general.

- **Facturación:** La tienda 4 tiene la más baja (\$10,383,757.00), lo que indica menor volumen de ventas.

- **Costo de envío:** Es el más bajo (\$234.59), lo que es una fortaleza; sin embargo, no compensa el bajo desempeño comercial general.

- **Producto más vendido:** Cama Box (62 unidades), una cifra competitiva pero no suficiente para revertir su baja facturación.

En comparación:

- **Tienda 1** destaca por su **alta facturación**, a pesar de tener la **peor calificación promedio**.
- **Tienda 2** mantiene un **rendimiento sólido y equilibrado** en todos los indicadores.
- **Tienda 3**, aunque con facturación media, tiene la **mejor calificación de clientes (4.05/5)** y buena rotación de productos.

### Conclusión

**La Tienda 4 no presenta una ventaja competitiva clara frente a las demás.** Entonces vender esta tienda permitiría al Sr. Juan liberar recursos financieros y operativos que podrían ser reinvertidos en un nuevo proyecto con mejores perspectivas de crecimiento y retorno.
